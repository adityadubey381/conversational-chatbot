# Cleaning the Dataset

```python
import re

def clean_text(example):
    text = example["text"]

    # remove special tokens
    text = text.replace("<|start|>", "")
    text = text.replace("<|end|>", "")
    text = text.replace("</s>", "")

    # remove special tokens
    text = re.sub(r"<\|.*?\|>", "", text)
    text = text.replace("</s>", "")

    # fix contractions (strong version)
    text = re.sub(r"\b(\w+)\s+'\s*(\w+)", r"\1'\2", text)
    text = re.sub(r"\b(\w+)\s+n['’]t\b", r"\1n't", text)

    # fix punctuation spacing
    text = re.sub(r"\s+([?.!,])", r"\1", text)
    text = re.sub(r"([?.!])([A-Za-z])", r"\1 \2", text)

    # normalize spaces
    text = re.sub(r"\s+", " ", text)

    return {"clean_text": text.strip()}
```

```python
def format_dialogue(example):
    text = example["clean_text"]

    # force newline before speakers
    text = re.sub(r"\s*A:", "\nA:", text)
    text = re.sub(r"\s*B:", "\nB:", text)

    # remove extra spaces/newlines
    text = re.sub(r"\n+", "\n", text)

    return {"formatted_text": text.strip()}
```

```python
train_dataset = train_dataset.map(clean_text)
train_dataset = train_dataset.map(format_dialogue)

val_dataset = ds["val"].map(clean_text)
val_dataset = val_dataset.map(format_dialogue)

test_dataset = ds["test"].map(clean_text)
test_dataset = test_dataset.map(format_dialogue)
```

# Creating a input and output pairs for the mode to teach the pattenr of replay

```python
def make_context_pairs(example):
    lines = example["formatted_text"].split("\n")

    inputs = []
    outputs = []

    for i in range(len(lines) - 1):
        if lines[i].startswith("A:") and lines[i+1].startswith("B:"):

            context_parts = []

            #taking last 2 sentences
            if i >= 2:
                context_parts = [lines[i-2], lines[i-1], lines[i]]
            elif i == 1:
                context_parts = [lines[i-1], lines[i]]
            else:
                context_parts = [lines[i]]

            # join with newline (IMPORTANT)
            context = "\n".join(context_parts)

            inputs.append(context)
            outputs.append(lines[i+1])

    return {"input": inputs, "output": outputs}
```

```python
# Data Collator
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=False   # VERY IMPORTANT
)
```

```python
from transformers import TrainingArguments, Trainer, DataCollatorWithPadding
import torch

training_args = TrainingArguments(
    output_dir="./chatbot",
    per_device_train_batch_size=4,
    num_train_epochs=2,
    logging_steps=100,
    save_steps=500,
    learning_rate=5e-5,
    fp16=torch.cuda.is_available(),   # if GPU supports
    report_to="none"
)
```

```python
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    data_collator=data_collator   # correct
)
```

```python
chat_history = ""

def generate_reply(user_input):
    global chat_history

    chat_history += f"User: {user_input}\nBot:"

    inputs = tokenizer(
        chat_history,
        return_tensors="pt",
        padding=True,
        truncation=True
    )

    output = model.generate(
        input_ids=inputs["input_ids"],
        attention_mask=inputs["attention_mask"],  # FIX warning
        max_length=300,
        pad_token_id=tokenizer.eos_token_id,
        do_sample=True,
        top_k=40,
        top_p=0.9,
        temperature=0.7,
        repetition_penalty=1.3,       # reduce repetition
        #no_repeat_ngram_size=2 # important
        min_length=20
    )

    decoded = tokenizer.decode(output[0], skip_special_tokens=True)

    # extract only bot reply
    reply = decoded[len(chat_history):]

    chat_history += reply + "\n"

    return reply.strip()
```
